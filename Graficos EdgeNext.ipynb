```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
FIGURES_PATH = "figures/"
```

```python
# Cargar csv
df = pd.read_csv("models/edgenext_last_layer/training_history.csv")
```

```python
df.head()
```

## 1. Plotear funcion de perdida 

```python
sns.set(style="whitegrid", font_scale=1.2)

# Graficar
plt.figure(figsize=(8, 5))
sns.lineplot(x=df.index + 1, y="train_loss", data=df, label="Loss conjunto entrenamiento")
sns.lineplot(x=df.index + 1, y="val_loss", data=df, label="Loss conjunto validacin")
plt.xlabel("Epoca")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
FIGURE_NAME = "loss_curve.pdf"
SAVE_FIG_PATH = os.path.join(FIGURES_PATH, FIGURE_NAME)
plt.savefig(SAVE_FIG_PATH)
```

## 2. Plotear PrAUC

```python
sns.set(style="whitegrid", font_scale=1.2)

# Graficar
plt.figure(figsize=(8, 5))
sns.lineplot(x=df.index + 1, y="train_pr_auc", data=df, label="PrAUC conjunto entrenamiento")
sns.lineplot(x=df.index + 1, y="val_pr_auc", data=df, label="PrAUC conjunto validacin")
plt.xlabel("Epoca")
plt.ylabel("PrAUC")
plt.legend()
plt.tight_layout()
FIGURE_NAME = "pr_auc_edgenext.pdf"
SAVE_FIG_PATH = os.path.join(FIGURES_PATH, FIGURE_NAME)
plt.savefig(SAVE_FIG_PATH)
```
